# Data Import, Exploration, and Cleaning

## ADNIMERGE Table from ADNI
- Datasource [ADNI website](http://adni.loni.usc.edu/data-samples/access-data/)
- Documentation 
> Data used in the preparation of this article were obtained from the Alzheimer’s Disease
Neuroimaging Initiative (ADNI) database (adni.loni.usc.edu). The ADNI was launched in
2003 as a public-private partnership, led by Principal Investigator Michael W. Weiner,
MD. The primary goal of ADNI has been to test whether serial magnetic resonance imaging
(MRI), positron emission tomography (PET), other biological markers, and clinical and
neuropsychological assessment can be combined to measure the progression of mild
cognitive impairment (MCI) and early Alzheimer’s disease (AD). For up-to-date information,
see www.adni-info.org.

```python
# import the relevant packages

import pandas as pd
import matplotlib.pyplot as plt

# run cell magics
%matplotlib inline
```

```python
# starting with object dtype to avoid errors in data consistency

adni_full = pd.read_csv('ADNIMERGE.csv', dtype='object')
```

```python
# research online for logical order of categories

DX = ['CN', 'MCI', 'Dementia']
```

```python
adni_full.DX_bl.value_counts()
```

```python
# research online for logical order of categories

DX_bl = ['CN', 'SMC', 'EMCI', 'LMCI', 'AD']
```

```python
# initialize empty dataframe

adni = pd.DataFrame()
```

```python
# set important datatypes

adni['RID'] = adni_full.RID.astype('int')
adni['EXAMDATE'] = pd.to_datetime(adni_full['EXAMDATE'])
adni['EXAMDATE_bl'] = pd.to_datetime(adni_full['EXAMDATE_bl'])
adni['Month'] = adni_full.Month.astype('int')
adni['PTGENDER'] = pd.Categorical(adni_full.PTGENDER)
adni['DX'] = pd.Categorical(adni_full.DX, ordered=True, categories=DX)
adni['DX_bl'] = pd.Categorical(adni_full.DX_bl, ordered=True, categories=DX_bl)
adni['PTEDUCAT'] = adni_full.PTEDUCAT.astype('int')
```

```python
# can't convert to floats  ['ABETA', 'TAU', 'PTAU']
# create a list of values that should be float dtype for quick looping

floats = ['AGE', 'CDRSB', 'ADAS11', 'ADAS13', 'MMSE', 'RAVLT_immediate', 'Hippocampus', 
          'Ventricles', 'WholeBrain', 'Entorhinal', 'MidTemp', 'FDG', 'AV45']
```

```python
# use list to convert dtypes to float and move to new df

for i in floats:
    adni[i] = adni_full[i].astype('float')
    
    # age has no baseline '_bl' equivalent
    if i == 'AGE':
        continue
    
    # every other column has a '_bl' equivalent to convert as well
    else:    
        y = i + '_bl'
        adni[y] = adni_full[y].astype('float')
```

```python
# examine the new dataframe

# comment out to save space
# adni.info
```

```python
cols = list(adni.columns.values)
```

```python
# generate NaN report

# add column names
nan_dict = {'column_name': list(cols)}

# initialze temp lists for looping
pct_list = []
sum_list = []

# generate a list of the pct null values
for col in cols:
    pct_list.append(adni[col].isnull().mean())
    sum_list.append(adni[col].isnull().sum())
    

# assign lists to the dictionary
nan_dict['pct_missing'] = pct_list
nan_dict['sum_missing'] = sum_list

# convert to dataframe
nan_report = pd.DataFrame(nan_dict)
```

```python
# isolate clinical data

clinical = adni_dx.loc[:, 'EXAMDATE_bl':'RAVLT_immediate_bl']
```

```python
clinical.shape
```

```python
# check the number of rows (visits) with complete clinical data

clinical_rmv = clinical.dropna(how='any')
```

```python
clinical_rmv.shape
```

```python
# only include patients with multiple visits that have complete data

# generate a count of exams for each patient
num_clin_exams = clinical_rmv.groupby('RID')['EXAMDATE_bl'].count()

# reduce this list to patients with multiple visits
clin_filter = num_clin_exams[num_clin_exams > 1]

# filter the clinical data to these patients
clin_data = clinical_rmv.loc[clin_filter.index]
```

```python
clin_data.shape
```

```python
# 9072 visits where the patient has multiple visits, with no missing data

clin_data.head()
```

```python
# create a dataframe of scans and check for complete data

# include only the scan data
scans = adni_dx.loc[:, 'Hippocampus':]

# add EXAMDATE_bl back in
scans['EXAMDATE_bl'] = adni_dx.EXAMDATE_bl

# filter for complete rows
scans_rmv = scans.dropna(how='any')
```

```python
scans_rmv.shape
```

```python
# down to 5006 rows, check for the number with multiple visits

# generate a count of exams for each patient
num_scan_exams = scans_rmv.groupby('RID')['EXAMDATE_bl'].count()

# reduce this list to patients with multiple visits
scan_filter = num_scan_exams[num_scan_exams > 1]

# filter the clinical data to these patients
scan_data = scans_rmv.loc[scan_filter.index]
```
