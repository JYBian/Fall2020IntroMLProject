#### 3. Which biomarkers change as a patient progresses into Alzheimer's Disease?

Knowing which measurements tend to change the most during the progression of AD could help to more quickly identify and diagnose AD. The idea behind this question to help identify when a patient may be experience a change in diagnosis. In other words, these measurements may help to more easily and quickly identify patients that are getting worse. The approach to answer this question involved calculating the change in each biomarker and adding that information to the `final_exam` dataframe. The plots below show the change in specific biomarkers grouped by change in diagnosis.

The approach to answer this question began by trying to identify variables that had the potential to answer this question. Two functions were created to easily reproduce two types of plots to search for this information. The first sets up parameters to feed into the plotting function, including breaking up the data into diagnosis groups. The groups used focused on change in diagnosis: No Change (916 patients), CN to MCI (37 patients), MCI to AD (176 patients), and CN to AD (5 patients).

```python
# calculate the changes for all variables between final and baseline and add to final_exam dataframe
# these calculated columns can be supplied to the plot functions below
eda.calc_deltas(final_exam)

# setup the parameters needed for plot_hist and plot_dist functions
dist_groups, labels, dist_bins = eda.setup_dist_plots(final_exam)
```

#### CDRSB Change: better at predicting AD than baseline values alone
While it appeared that the baseline CDRSB values had little predictive power, the plots below seem to suggest that changes in CDRSB of about 2.5 or more are indicitave of a progression to AD.

```python
# plot the CDRSB_delta scores

_ = eda.plot_histdist('CDRSB_delta', dist_groups, labels, dist_bins)
```

#### ADAS11 changes suggest progression to AD
The plots below show that patients that experienced an ADAS11 change of 10 or more saw a diagnosis change to AD in most cases. 

```python
# plot the ADAS11_delta scores

_ = eda.plot_histdist('ADAS11_delta', dist_groups, labels, dist_bins)
```

#### Changes in ADAS13 scores, like ADAS11, suggest a change in diagnosis
It appears that changes in ADAS13 scores can also indicate a progression to AD, with a threshold value of a change in ADAS13 of about 11-12.

```python
# plot the ADAS13_delta scores

_ = eda.plot_histdist('ADAS13_delta', dist_groups, labels, dist_bins)
```

#### Changes in MMSE are associated with progression to AD
The plots below suggest that a change in MMSE of -3 or more should be of concern. Any amount of change greater than that is likely indicative of AD.

```python
# plot the MMSE_delta scores

_ = eda.plot_histdist('MMSE_delta', dist_groups, labels, dist_bins)
```

#### Changes in RAVLT_immediate less obvious than other measures
It appears that a change in RAVLT_immediate of -18 or more should be concerning, there is more overlap with the 'No Change' group for this measure than some of the other cognitive tests. A change in RAVLT_immediate should be more associated with short term variables such as mood or other temporary factors, as suggested in AD literature.

```python
# plot the RAVLT_immediate_delta scores

_ = eda.plot_histdist('RAVLT_delta', dist_groups, labels, dist_bins)
```

#### Hippocampus measurements show marked decline with progression to AD
A change in hippocampus volume of about -750 or more is largely indicative of progression towards AD. There were very few patients that displayed no change in diagnosis that had changes in Hippocampus volume greater than -750.

```python
# plot the Hippocampus_delta scores

_ = eda.plot_histdist('Hippocampus_delta', dist_groups, labels, dist_bins)
```

#### Ventricle volume increases with progression to AD
An increase in ventricles of 10000 or more is highly suggestive of progression to AD. There are some patients from the 'No Change' group with ventricle changes over 10000. Further exploration (calculated below the graph) shows that only about 15.8% of patients with a ventricle change of 10000 or more where diagnosed CN at the end of the study.

```python
# plot the Ventricles_delta scores

_ = eda.plot_histdist('Ventricles_delta', dist_groups, labels, dist_bins)
```

```python
# 215 patients with ventricles_delta >= 100000, only 34 were 'CN' at final diagnosis (about 15.8%)

print(final_exam[final_exam.Ventricles_delta >= 10000].shape[0],
final_exam[(final_exam.Ventricles_delta >= 10000) & (final_exam.DX == 'CN')].shape[0])
```

```python
# what percentage of patients had ventricle change of 10000 or more and were CN?

print(34/215 * 100)
```

#### Only large WholeBrain changes suggest progression to AD, when considered alone
WholeBrain decrease of 100000 or more suggest progression to AD. Only 22 patients had changes this great, so this measurement is likely most useful when used in conjunction with other data to identify individuals progressing to AD.

```python
# plot the WholeBrain_delta scores

_ = eda.plot_histdist('WholeBrain_delta', dist_groups, labels, dist_bins)
```

```python
# how many patients actually had changes in WholeBrain of -100000 or more?

final_exam[final_exam.WholeBrain_delta <= -100000].shape[0]
```

#### Changes in Entorhinal of -1000 or greater suggest progression to AD
Most patients that experienced entorhinal decrease of -1000 or more saw a change in diagnosis to AD. This measurement could be useful in identifying progression of the disease, especially in conjunction with other biomarkers.

```python
# plot the Entorhinal_delta scores

_ = eda.plot_histdist('Entorhinal_delta', dist_groups, labels, dist_bins)
```

#### Changes in MidTemp also suggest progression to AD
Most patients that experienced a MidTemp decrease of -3000 or more progressed to AD. As all of the groups that experienced a change in diagnosis have large sections of their distributions to the left of the No Change group, the hypothesis is that MidTemp values are one of the most indicative biomarkers for indicating progression of AD.

```python
# plot the MidTemp_delta scores

_ = eda.plot_histdist('MidTemp_delta', dist_groups, labels, dist_bins)
```

## 4. Which baseline measurements might be able to predict a final diagnosis of Alzheimer's disease?

This is an extremely important question to try and answer. To begin this investigation, the data were separated by gender, then patients were categorized by their final diagnosis. All of the different biomarkers were analyzed, but only the most interesting and most promising variables are displayed below.

```python
# setup the indicators for histograms comparing males and females
# this divides up the data into groups based on final diagnosis and gender

setup_indicators = eda.setup_indicators(final_exam)
```

#### Some baseline ADAS11 and ADAS13 might be able to predict Alzheimer's diagnosis
The plots below show that ADAS11 and ADAS13 distributions of patients with AD stand out to the right from the other distributions. These plots do not take into account those patients that already had AD at the beginning of the study, so those values was calculated for males and females based on the thresholds indicated in the plots below.

#### Females:
#### ADAS11 Baseline threshold of 15-16 predicts AD
- There were 40 females who had ADAS11 scores over 15 that didn't have an initial diagnosis of AD. Of those 40, 28 of them were later diagnosed with AD. This suggests that a baseline ADAS11 of 15 or higher shows a relatively high risk of developing AD in patients that don't already have the disease. When the threshold is increased to 16, 25/33 females with scores that high that didn't already have AD, were diagnosed with AD by the end of the study.

#### ADAS13 Baseline threshold of 25 predicts AD
- For ADAS13 24/30 female patients that didn't already have an AD diagnosis that scored 25 or higher on ADAS13 were later diagnosed with AD.

#### Males:
#### ADAS11 Baseline threshold not great at predicting AD alone
- For ADAS11, only 30/55 male patients with baseline ADAS11 over 15 ended up with AD diagnosis. This is a prediction rate of just 54.55%.

#### ADAS13 Baseline threshold of 25 does ok at predicting AD
- For ADAS13, about 61% of males with baseline of 25 or higher that didn't have AD ended with an AD diagnosis.

```python
# plot baseline ADAS11 and ADAS13 scores separated by gender and final diagnosis

_ = eda.plot_indicators(column1='ADAS11_bl', column2='ADAS13_bl', si=setup_indicators)
```

```python
# summarize female ADAS11 baseline threshold

eda.summarize_bl_thresholds(final_exam, column='ADAS11_bl', gender='Female', threshold=16)
```

```python
# summarize female ADAS13 baseline threshold

eda.summarize_bl_thresholds(final_exam, column='ADAS13_bl', gender='Female', threshold=25)
```

```python
# summarize male ADAS11 baseline threshold

eda.summarize_bl_thresholds(final_exam, column='ADAS11_bl', gender='Male', threshold=15)
```

```python
# summarize male ADAS13 baseline threshold

eda.summarize_bl_thresholds(final_exam, column='ADAS13_bl', gender='Male', threshold=25)
```

## Summary of Initial Findings

#### Several Biomarkers Identified as Potential Predictors of Alzheimer's Disease
Even though only 218 patients had a diagnosis progress towards AD during the study (with 172 already having AD) several biomarkers have been identified that may be able to predict AD. 

#### Watch Variables for Changes
A series of histogram and distribution plots showed that most of the biomarkers likely have a certain amount of change that should raise concerns. Further statistical analysis is needed to explore what amount of change is associated with a change in diagnosis.

#### Four Baseline Measures Have Some Predictive Power
When the data were separated into groups based on gender, then futher subdivided into final diagnosis, there were four variables that seemed to hold some predictive power for AD. ADAS11 and ADAS13 were the best cognitive tests that appeared to hold value, while Hippocampus and MidTemp measurements also had threshold values that could suggest a patient is at risk for developing AD.

#### Further Analysis is Needed
Some promising insights were gained including the identification of four measures that could be used to predict AD in patients that have not yet been diagnosed with the disease. This raises questions about what exactly the threshold values would be, are there differences between men and women as the data seem to suggest, and is some combination of those variables more powerful at predicting the disease than any single biomarker alone?
